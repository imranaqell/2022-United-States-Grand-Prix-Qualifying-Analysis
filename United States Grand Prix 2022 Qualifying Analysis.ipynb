```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
session = ff1.get_session(2022, 'Austin', 'Q')
session.load()
```

```python
sai_lap = session.laps.pick_driver('SAI').pick_fastest()
ver_lap = session.laps.pick_driver('VER').pick_fastest()
lec_lap = session.laps.pick_driver('LEC').pick_fastest()

sai_lap = sai_lap.get_car_data().add_distance()
ver_lap = ver_lap.get_car_data().add_distance()
lec_lap = lec_lap.get_car_data().add_distance()
```

```python
#Sainz VS Leclerc
plotting.setup_mpl()

fig, ax = plt.subplots(3, figsize = (9, 16))
fig.suptitle("SAI VS LEC United States GP Qualifying Lap Comparison", y=0.895)

ax[0].plot(sai_lap['Distance'], sai_lap['Speed'], color='#A10000', label='Sainz')
ax[0].plot(lec_lap['Distance'], lec_lap['Speed'], color='#FCE205', label='Leclerc')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(sai_lap['Distance'], sai_lap['Throttle'], color='#A10000', label='Sainz')
ax[1].plot(lec_lap['Distance'], lec_lap['Throttle'], color='#FCE205', label='Leclerc')
ax[1].set(ylabel='Throttle')

ax[2].plot(sai_lap['Distance'], sai_lap['Brake'], color='#A10000', label='Sainz')
ax[2].plot(lec_lap['Distance'], lec_lap['Brake'], color='#FCE205', label='Leclerc')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
#Sainz VS Verstappen
fig, ax = plt.subplots(3, figsize = (9, 16))
fig.suptitle("SAI VS VER United States GP Qualifying Lap Comparison", y=0.895)

ax[0].plot(sai_lap['Distance'], sai_lap['Speed'], color='#A10000', label='Sainz')
ax[0].plot(ver_lap['Distance'], ver_lap['Speed'], color='#1034A6', label='Verstappen')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(sai_lap['Distance'], sai_lap['Throttle'], color='#A10000', label='Sainz')
ax[1].plot(ver_lap['Distance'], ver_lap['Throttle'], color='#1034A6', label='Verstappen')
ax[1].set(ylabel='Throttle')

ax[2].plot(sai_lap['Distance'], sai_lap['Brake'], color='#A10000', label='Sainz')
ax[2].plot(ver_lap['Distance'], ver_lap['Brake'], color='#1034A6', label='Verstappen')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```
